# Credit Risk Analysis — Business Insights Report

**Dataset:** Home Credit Default Risk — 307,511 loan applications  
**Model:** XGBoost Classifier with SHAP Explainability  
**Tools:** Python, Scikit-learn, XGBoost, SHAP, Imbalanced-learn  
**Author:** M. Rijal Alghafiqih

## Why I Built This Project

Credit risk modeling is the backbone of every lending business. Every
time a bank or fintech company approves a loan, they are making a bet
that the borrower will pay them back. Get it wrong too many times and
the business fails. Get it too conservative and you turn away good
customers and lose revenue.

I built this project to demonstrate end-to-end credit risk modeling
using real loan application data from Home Credit, a fintech company
operating across Southeast Asia and Eastern Europe. The goal was not
just to build an accurate model but to build one that can be explained
to regulators, loan officers, and customers — because in the real world,
a black box model is not enough.

## What I Found — 5 Key Insights

### Insight 1: External credit scores are the strongest predictor of default

The SHAP analysis confirmed what credit analysts have known for decades
— external credit scores (EXT_SOURCE_2 and EXT_SOURCE_3) are the most
predictive features in the model. Applicants with scores below 0.3 are
significantly more likely to default than those above 0.6.

What surprised me was that income type and employment sector appeared
almost as important as the credit scores themselves. This suggests that
for this specific lender's customer base, behavioral and demographic
factors carry nearly as much weight as formal credit history.

The business implication is clear: lenders should prioritize enriching
their data with multiple external credit bureau scores rather than
relying on a single score, especially for applicants with thin credit
files.

### Insight 2: Young applicants are disproportionately risky

The EDA revealed that applicants under 25 default at 12.3% — nearly
2.5 times the rate of applicants over 55 who default at only 5.1%.
This age gap is one of the most actionable findings in the analysis.

However, rejecting all young applicants would be both legally problematic
and commercially shortsighted. Young borrowers who repay successfully
become loyal long-term customers. A better strategy is to apply stricter
loan-to-income limits for applicants under 30 while offering smaller
initial credit limits that can be increased as the relationship matures.

### Insight 3: Cash loans carry higher risk than revolving loans

Cash loans default at 8.5% versus 5.5% for revolving loans. This makes
intuitive sense — revolving credit like credit cards gives borrowers
flexibility to manage cash flow, while a fixed cash loan creates a rigid
monthly obligation that is harder to service during financial stress.

Lenders should consider this when pricing products. Cash loans to
higher-risk demographic segments should carry higher interest rates
or require additional collateral to compensate for the elevated default
risk.

### Insight 4: Employment stability matters more than income level

I expected income to be the strongest financial predictor but the data
told a different story. Applicants with less than 2 years of employment
showed significantly higher default rates regardless of income level.
Someone earning a high salary who just started a new job is actually
riskier than a moderate-income applicant who has been stably employed
for 5 years.

This insight is practically useful for underwriting decisions. Employment
duration should be a hard filter in the application process, with
additional scrutiny for applicants with less than 12 months at their
current employer.

### Insight 5: The model can explain individual decisions

The SHAP waterfall analysis demonstrated that every prediction can be
decomposed into specific, auditable contributions from individual features.
For the high-risk applicant I analyzed, the model flagged them primarily
due to low external credit scores combined with working-class income type
and single marital status.

This level of explainability means the model is not just a black box
score — it can generate a specific reason for every credit decision.
This is essential for regulatory compliance under fair lending laws and
for giving applicants meaningful feedback about why their application
was declined.

## Model Performance Summary

| Metric | Logistic Regression | Random Forest | XGBoost |
|--------|-------------------|---------------|---------|
| AUC-ROC | 0.7422 | 0.6793 | 0.7488 |
| Rank | 2nd | 3rd | 1st |

XGBoost was selected as the final model for three reasons. First, it
achieved the highest AUC-ROC of 0.7488 which means it correctly
distinguishes between defaulters and non-defaulters 74.88% of the time.
Second, it handles class imbalance natively through scale_pos_weight.
Third, it integrates seamlessly with SHAP for regulatory explainability.

An AUC of 0.75 is considered solid for consumer credit risk modeling.
Industry benchmarks typically range from 0.70 for simple models to 0.85
for models with rich behavioral data like transaction history and payment
patterns. Our model achieved this without any behavioral data, using only
the application-level features provided at the time of loan request.

## Class Imbalance — The Hidden Challenge

One of the most important technical decisions in this project was how to
handle the 8% default rate. Without intervention, any model trained on
this data would be heavily biased toward predicting no default because
that is the correct answer 92% of the time.

I used SMOTE (Synthetic Minority Oversampling Technique) on the training
data to create synthetic default cases and bring the minority class up
to 30% of the training set. This gave the model enough examples of
default behavior to learn meaningful patterns.

Critically, I never applied SMOTE to the test data. Doing so would
inflate the evaluation metrics and give a false picture of real-world
performance. The test metrics are based purely on the original
imbalanced distribution, which is what the model will actually encounter
in production.

## Limitations and Next Steps

This model has real limitations that I want to be transparent about.

First, the dataset is from Eastern European and Southeast Asian markets.
A model trained on this data may not transfer directly to other markets
with different credit cultures, regulatory environments, and economic
conditions.

Second, the model only uses application-level features. In a real
production environment, behavioral data like transaction patterns,
payment history, and spending behavior would significantly improve
performance. Models with this type of data routinely achieve AUC above
0.80.

Third, the model requires regular retraining as economic conditions
change. A model trained on pre-2020 data would have struggled to
predict defaults during the COVID period because the underlying
relationship between features and default risk shifted dramatically.

If I were to extend this project the next steps would be to incorporate
bureau data for richer credit history, build a monitoring pipeline to
detect model drift over time, and deploy the model as a REST API that
returns both a probability score and a SHAP-based explanation for
each decision.

## Conclusion

This project demonstrates that credit risk modeling requires more than
just picking an algorithm and running it. The real work is in understanding
the business problem, handling the class imbalance honestly, choosing
the right evaluation metrics, and being able to explain every decision
the model makes.

The XGBoost model with AUC 0.7488 is a solid baseline that could be
deployed in a real lending environment with proper monitoring and
regulatory review. More importantly, the SHAP explainability layer means
every credit decision can be audited and explained — which is the
difference between a model that sits in a notebook and one that actually
gets used in production.
